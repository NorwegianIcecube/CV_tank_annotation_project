```python
import cv2
import numpy as np
import os
```

```python
def draw_annotations(image_path, annotation_path, output_path):
    # Read the image
    img = cv2.imread(image_path)

    # Check if image was loaded successfully
    if img is None:
        print("Error: Unable to load image")
        return

    # Get image width and height
    img_h, img_w = img.shape[:2]

    # Loop through each annotation
    with open(annotation_path, 'r') as f:
        annotations = f.readlines()

        for annotation in annotations:
            # Split annotation into individual values
            values = annotation.split()

            # Extract class ID
            class_id = int(values[0])

            # Extract polygon vertices (x, y coordinates)
            vertices = [(float(values[i]) * img_w, float(values[i + 1]) * img_h) for i in range(1, len(values), 2)]
            vertices = np.array(vertices, dtype=np.int32).reshape((-1, 1, 2))

            # Print class ID and vertices for debugging
            print("Class ID:", class_id)
            print("Vertices:", vertices)
            
            # Draw polygon on image
            cv2.polylines(img, [vertices], isClosed=True, color=(0, 255, 0), thickness=2)
            
    # Save the modified image
    cv2.imwrite(output_path, img)

# Example usage
a = os.listdir('leopard_frames')
b = os.listdir('leopard_frames_a_annotate')
#image_path = './tank_frames/frame_10.jpg'
#annotation_path = './tank_frames/frame_10.txt'
output_path = 'leopard_autoimg'

for i in b:
    print(b)
    #draw_annotations(a, b, output_path)
```
